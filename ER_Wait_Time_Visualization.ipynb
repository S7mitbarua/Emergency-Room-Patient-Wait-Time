```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulate data for emergency room wait times
np.random.seed(42)  # For reproducibility

days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
hours_of_day = list(range(24))

data = {
    'Day': np.random.choice(days_of_week, size=1000),
    'Hour': np.random.choice(hours_of_day, size=1000),
    'WaitTime': np.random.randint(0, 120, size=1000)  # Simulated wait time in minutes
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Hour', y='WaitTime', data=df, hue='Day', ci=None)
plt.title('Emergency Room Wait Times - Daily Trends')
plt.xlabel('Hour of the Day')
plt.ylabel('Wait Time (minutes)')
plt.legend(title='Day')
plt.show()
```
